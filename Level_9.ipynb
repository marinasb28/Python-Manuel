# <font color="magenta">RETO 4: El módulo random</font>

Haz un juego que genera un número aleatorio del 1 al 100 y se trata
de que lo adivines.

El juego te pide que introduzcas números y te indica si el número introducido
es mayor o menor que el número que hay que adivinar.

Tienes 7 intentos, si no, el programa se termina y te dice que has perdido.

```
guess_num = random.randint(1,100) # va de 1 a 100
print("El número que hay que averiguar es: {}".format(guess_num))
intentos = 0
jugando = True
while jugando:
  intentos += 1
  if intentos <= 7:
    guess = int(input("Write a number: "))
    if guess == guess_num:
      print("Has acertado!")
      jugando = False
    elif guess > guess_num:
      print("El número es menor.")
    elif guess < guess_num:
      print("El número es mayor.")
  else:
    print("No tienes más intentos.Has perdido!")
    jugando = False

  
```

# <font color="magenta">RETO 5: Adivina el número</font>

Se trata de mejorar el juego que acabamos de hacer "Adivina el número".
Para ello, utilizando la función choice, hay que hacer que el ordenador
muestre una frase diferente cada vez que se elige un número.

```
guess_num = random.randint(1,100) # va de 1 a 100
print("El número que hay que averiguar es: {}".format(guess_num))
intentos = 0
jugando = True
bajo = ["Muy bajo","Por los suelos","Baja de las nubes"]
alto = ["Muy alto","Hasta el infinito y más allá","Coge la escalera y sube"]
while jugando:
  intentos += 1
  if intentos <= 7:
    guess = int(input("Write a number: "))
    if guess == guess_num:
      print("Has acertado!")
      jugando = False
    elif guess > guess_num:
      print(random.choice(bajo))
    elif guess < guess_num:
      print(random.choice(alto))
  else:
    print("No tienes más intentos.Has perdido!")
    jugando = False

  
```

Haz un juego que sea un concurso matemático, en el que te aparecen sumas y
multiplicaciones entre el 1 y el 9 generadas al azar por el ordenador.

Cada vez que aciertas una operación te da un punto.

Hay que implementar un tiempo de 10 segundos, de tal forma que en cuanto
pasen los 10 segundos se acabe el juego y te muestre los puntos.

```
print("\r\tCONCURSO MATEMÁTICO")
print("Realiza todas las operaciones posible en 10s.")
comienzo = input("Pulsa ENTER para empezar: ")
puntos = 0
total = 0
operaciones = ["+","-","*"]
jugando = True
inicio = time.time()
while jugando:
  num1 = random.randint(1,10)
  num2 = random.randint(1,10)
  op = random.choice(operaciones)
  if comienzo =="":
    start = time.time() #aquí comenzamos a contar
  if op == "+":
    resultado = int(input("{}+{}= ".format(num1,num2)))
    if resultado == num1+num2:
      puntos += 1
    else:
      print("Te has confundido.")
  elif op == "-":
    resultado = int(input("{}-{}= ".format(num1,num2)))
    if resultado == num1-num2:
      puntos += 1
    else:
      print("Te has confundido.")
  elif op == "*":
    resultado = int(input("{}*{}= ".format(num1,num2)))
    if resultado == num1*num2:
      puntos += 1
    else:
      print("Te has confundido.")
  final = time.time()
  total += 1
  if final-inicio >= 10:
    print("Se acabó el tiempo.")
    jugando = False
print("Has ganado {}/{} puntos.".format(puntos,total))
```

Haz un programa que lleve a cabo una cuenta atrás de 10 segundos.

1. Que muestre cada segundo. (10, 9, 8, ...)
2. Que muestre cada 5 décimas de segundo. (10, 9.5, 9, 8.5, ...)

Mide el tiempo para comprobar que efectivamente transcurren unos 10 segundos

```
# Que muestre cada segundo. (10, 9, 8, ...)
inicio = time.time()
for x in range(10,-1,-1):
  print(x)
  time.sleep(1) 
final = time.time()
print("Tiempo empleado: {}".format(round(final-inicio-1,2)))
```

# <font color="magenta">RETO 12: Cuenta atrás</font>

Haz un programa que calcule la suma de los números del 1 al 1.000.000. 

Hazlo con el bucle while y con el bucle for. 

Utiliza la función time.perf_counter() , o la función time.clock() para comprobar el tiempo que tarda en ejecutarse el programa con cada uno de los dos bucles.

```
inicio = time.perf_counter()
counter = 0
index = 1
while index <= 10**6:
  counter += index
  index += 1
final = time.perf_counter()
print(counter)
print(final-inicio)
```

```
inicio = time.perf_counter()
counter = 0
for x in range((10**6)+1):
  counter += x
final = time.perf_counter()
print(counter)
print(final-inicio)
```

Haz un programa es que un juego de memoria.
 
Se te muestran cuatro colores de forma aleatoria, y tras tres segundos, se te pide que les pongas en el orden en el que aparecían. 

Si se acierta se vuelven a mostrar los cuatro colores en otro orden aleatorio distinto.

```
colores = ["rojo","amarillo","verde","azul"]
random.shuffle(colores) #altera nuestra lista inicial
print(colores)
time.sleep(3)
for x in range(1,10):
  print("*"*x)
for x in range(10,-1,-1):
  print("*"*x)
puntos = 0
print("En qué orden aparecían?")
for x in range(len(colores)):
  result = input("Escribe el color {}: ".format(x+1))
  if result == colores[x]:
    print("Correcto.")
    puntos += 1
  else:
    print("Incorrecto")
print("{}/4".format(puntos))
```

# <font color="magenta">RETO 14: Barajando elementos</font>

Mejora el programa del juego de memoria que acabamos de hacer
mediante la función sample().

En vez de una lista de cuatro colores, utiliza una lista de 10 colores.

```
colores = ["rojo","amarillo","verde","azul","negro","lila","blanco","gris","rosa","marron"]
colores = random.sample(colores,4) #generamos lista nueva
print(colores)
time.sleep(3)
for x in range(1,10):
  print("*"*x)
for x in range(10,-1,-1):
  print("*"*x)
puntos = 0
print("En qué orden aparecían?")
for x in range(len(colores)):
  result = input("Escribe el color {}: ".format(x+1))
  if result == colores[x]:
    print("Correcto.")
    puntos += 1
  else:
    print("Incorrecto")
print("{}/4".format(puntos))
```
